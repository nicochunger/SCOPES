# Helper file to create plots for the documentation of SCOPES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def merit(x, start_time, start_time_tolerance, steepness):
    arg1 = (x - (start_time - start_time_tolerance)) / steepness
    arg2 = ((start_time + start_time_tolerance) - x) / steepness
    return np.tanh(arg1) + np.tanh(arg2)


x = np.linspace(-3, 3, 200)
plt.figure(figsize=(8, 5))
plt.plot(x, merit(x, 0, 1, 0.1), "k", lw=2, label=r"$\delta=1, \alpha = 0.1$")
plt.plot(x, merit(x, 0, 1, 0.3), "k--", lw=2, label=r"$\delta=1, \alpha = 0.3$")
plt.plot(x, merit(x, 0, 2, 0.1), "k:", lw=2, label=r"$\delta=2, \alpha = 0.1$")
plt.legend()
plt.xlabel("Time")
plt.ylabel("Merit")
plt.tight_layout()
plt.savefig("timecritical_merit.png", dpi=300)
```
